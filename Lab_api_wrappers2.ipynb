```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
your_client_id = "c4e142cbe4ef45xxxxxx"
your_client_secret = "dbfc42c52e9146xxxxxx"

sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=your_client_id,
                                                           client_secret=your_client_secret))

def get_playlist_tracks(username, playlist_id):

    results = sp.user_playlist_tracks(username, playlist_id)
    tracks = results['items']

    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])

    return tracks

tracks = get_playlist_tracks("spotify", "37i9dQZEVXbMDoHDwVN2tF")
tracks
```

```python
track_data = []
for track in tracks:
    track_info = {
        'Track Name': track['track']['name'],
        'Artist': ', '.join([artist['name'] for artist in track['track']['artists']]),
        'Album': track['track']['album']['name'],
        'Release Date': track['track']['album']['release_date'],
        'Popularity': track['track']['popularity']
    }
    track_data.append(track_info)
df = pd.DataFrame(track_data)
print(df)
```
